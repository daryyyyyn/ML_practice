```python

import csv
from pathlib import Path
```

```python
file_path = Path("Salaries.csv")


with file_path.open(mode="r", encoding="utf-8", newline="") as f:
    reader = csv.reader(f)

    header = next(reader)

    data_rows = []
    for i, row in enumerate(reader):
      if i < 10:
        data_rows.append(row)
      else:
        break

    print("data_rows ұзындығы:", len(data_rows))
    print("Бірінші элемент (бірінші қызметкер):", data_rows[0])
```

```python
header = ['Id', 'EmployeeName', 'JobTitle', 'BasePay', 'OvertimePay', 'OtherPay']
name_index = header.index('EmployeeName')

first_three = data_rows[:3]

first_three_names = tuple(row[name_index] for row in first_three)

print(first_three_names)
```

```python
title_index = header.index('JobTitle')
employee_titles = {row[name_index]: row[title_index] for row in first_three}

print(employee_titles)
```

```python
basepay_index = header.index('BasePay')
base_pays_str = [str(row[basepay_index]) for row in data_rows]

print(base_pays_str)
```

```python
title_index = header.index('JobTitle')

job_titles = []

for row in data_rows:
    job_titles.append(row[title_index])

unique_titles = set(job_titles)

print(job_titles)

print(unique_titles)
```
