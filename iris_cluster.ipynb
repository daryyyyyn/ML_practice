```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
X = iris.data
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

centers_pca = pca.transform(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', s=50, alpha=0.7, label='Деректер')

plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, marker='X', label='Центрлер')

plt.title("K-Means Кластерлеу (PCA арқылы визуализация)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X)
    inertia.append(km.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(k_range, inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Method (Оптималды K табу)')
plt.xlabel('Кластер саны (k)')
plt.ylabel('Inertia (Қателік)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score

models = {
    "K-Means (k=3)": KMeans(n_clusters=3, random_state=42, n_init=10),
    "Hierarchical (k=3)": AgglomerativeClustering(n_clusters=3),
    "DBSCAN (eps=0.5)": DBSCAN(eps=0.5, min_samples=5)
}

results = {}

for name, model in models.items():
    labels = model.fit_predict(X)
    if len(np.unique(labels)) > 1:
        score = silhouette_score(X, labels)
        results[name] = score
    else:
        results[name] = "Есептелмеді (Кластер аз)"

    print(f"{name:20} | Silhouette Score: {results[name]}")
```
